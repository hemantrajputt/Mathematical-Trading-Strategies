```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
indices = ['^FTSE', '^FCHI', '^IXIC', '^NYA', '^XAX']
equity = ['MSFT' , 'AAPL' , 'TSLA', 'AMD' , 'NVDA']
```

```python
data_frame = pd.DataFrame()
for equity in equity:
    data_frame_tmp=yf.download(equity, start='2010-01-01', end='2023-05-10')
    data_frame_tmp['Ticker'] = equity
    data_frame = pd.concat([data_frame, data_frame_tmp])
data_frame
```

```python
data_frame.reset_index(inplace=True)
```

```python
data_frame.groupby('Ticker')['Date'].agg(['min', 'max', 'count'])
```

```python
data_frame1 = data_frame[['Date', 'Ticker', 'Adj Close']]
# make columns names prettier
data_frame1.columns = ['date', 'ticker', 'price']
data_frame1_temp = data_frame1
data_frame1
```

```python
data_frame2 = data_frame1.pivot_table(index=['date'], columns='ticker', values=['price'])
data_frame2.columns = [col[1] for col in data_frame2.columns.values]
data_frame2
```

```python
daily_returns = data_frame2.pct_change()
daily_returns = daily_returns[1:]
daily_returns
```

```python
cum_daily_returns = (1 + daily_returns).cumprod() - 1
cum_daily_returns = cum_daily_returns.reset_index()
cum_daily_returns
```

```python
cum_return_entire_period = cum_daily_returns.iloc[:, 1:].tail(1)
cum_return_entire_period * 100
```

```python
DF_daily_returns = daily_returns.reset_index()
data_frame_4 = DF_daily_returns.melt(id_vars=['date'], var_name='ticker', value_name='daily_return')
data_frame_4['daily_return_pct'] = data_frame_4['daily_return'] * 100
data_frame_4
```

```python
data_frame5 = data_frame_4[['date', 'ticker', 'daily_return']]
data_frame5
```

```python
data_frame6 = data_frame5.pivot_table(index=['date'], columns='ticker', values=['daily_return'])
# flatten columns multi-index, `date` will become the dataframe index
data_frame6.columns = [col[1] for col in data_frame6.columns.values]
data_frame6
```

```python
import math
for column in data_frame6.columns[0:]:
    daily_volatility = data_frame6[column].std()
    print(column)
    print('Daily volatility: ', '{:.2f}%'.format(daily_volatility*100))
    monthly_volatility = math.sqrt(21) * daily_volatility *100
    print('Monthly volatility: ', '{:.2f}%'.format(monthly_volatility))
    annual_volatility = math.sqrt(252) * daily_volatility *100
    print('Annual volatility: ', '{:.2f}%'.format(annual_volatility))
```

```python
for column in data_frame6.columns[0:]:
    downside_returns = data_frame6.loc[data_frame6[column] < 0]
    expected_return = data_frame6[column].mean()
    down_stdev = downside_returns[column].std()
    rfr=0
    sortino_ratio = (expected_return - rfr)/down_stdev
    sortino_ratio=sortino_ratio*np.sqrt(252)
    print(column + " : ", '{:.6f}'.format(sortino_ratio))
```
